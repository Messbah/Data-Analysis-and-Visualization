# CO2 Calculation

```python
total_co2 = (((table['SMS Sent'].sum()*0.014)/1000)*52)/10
total_co2
```

```python
table['CO2 Emission 10 Weeks'] = ((table['SMS Sent']*0.014)/1000)
```

```python
table['CO2 Emission'] = (((table['SMS Sent']*0.014)/1000)*52)/10
```

```python
myplot = table['CO2 Emission 10 Weeks'].plot(figsize=(15, 10), kind='line', rot=90, color = "blue", title = "A Graph of Users against their CO2 Emission from sent SMS")
#myplot.figure(figsize=(3, 3))
myplot.set_xlabel("Students")
myplot.set_ylabel("Co2 Emission in 10 Weeks")
plt.savefig('user-wise emission from SMS.jpg')
```

```python
table['CO2 Emission 10 Weeks'].cumsum()
```

```python
import matplotlib.patches as mpatches
fig, (ax1) = plt.subplots(1)
fig.suptitle('CO2 Emission by SMS')

ax1.bar(table.index, table['CO2 Emission'],color="red")
ax1.bar(table.index, table['CO2 Emission 10 Weeks'], color="blue")
ax1.set_xlabel("Users")
ax1.set_ylabel("Co2 Emission in KG Every year")

red_patch = mpatches.Patch(color='red', label='Emission in a year(Prediction)')
blue_patch = mpatches.Patch(color='blue', label='Emission in 10 weeks')
plt.legend(handles=[red_patch,blue_patch])
plt.savefig('user-wise emission from SMS_weekVSyear.jpg')
```
