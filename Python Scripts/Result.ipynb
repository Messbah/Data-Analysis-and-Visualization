```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

%matplotlib inline
grades = pd.read_csv(r"D:\Lectures\England\DAV\Report\dataset\education\grades.csv")
grades.head()
```

```python
phone_usage = pd.read_csv(r'C:\Users\muhad\DAV\Level_2\phone_usage.csv')
phone_usage.head()
```

```python
final = pd.DataFrame()
```

```python
for i in phone_usage.index:
     for j in grades.index:
            if(phone_usage['User'][i]==grades['uid'][j]):
                #print(phone_usage['Phone Usage'][i])
                #print(grades[' gpa all'][j])
                x = pd.DataFrame(data={'User ID': grades['uid'][j], 'Phone Usage':phone_usage['Phone Usage'][i], 'GPA': grades[' gpa all'][j]}, index=[0])
                final = final.append(x)
                
                
```
